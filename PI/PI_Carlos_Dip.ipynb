Escolha 1: Professor de GDE <-> Gradient Descent

Para machine learning, existem várias técnicas matemáticas de treinamento. Uma delas é conhecida como Gradient Descent, ou descida de gradiente. Essa técninca consistem em utilizar uma operação especial de vetores, para determinar a direção na qual devemos mudar nossos parâmetros para obter resultados melhores. Semelhante à ideia de soltar uma bolinha numa bacia, que ao longo do tempo acaba se acomodando num ponto de mínimo da bacia (no caso o fundo). É comparável também a ideias de oferta e demanda, onde existe essa noção de equilíbrio, já que para os consumidores é desejável o preço mínimo, o que faz com que a demanda caia conforme os preços sobem. A semelhança com o gradient descent está nessa noção de equilíbrio. É desejável que o modelo se ajuste aos dados, para ser capaz de prever novos eventos ou simplesmente explicar um fenômeno, porêm dentre as informações disponíveis para fazer esse treinamento, é necessária a busca por um equilíbrio, de forma que seja possível levá-las todas em conta, mas não demasiadamente.

## Modelagem

```python
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler, PolynomialFeatures
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.metrics import mean_squared_error

all_data
```

```python
state, year = ['STATE'], ['YEAR']
numeric = [ 'ENROLL',
            'TOTAL_REVENUE',
            'FEDERAL_REVENUE',
            'STATE_REVENUE',
            'LOCAL_REVENUE',
            'TOTAL_EXPENDITURE',
            'INSTRUCTION_EXPENDITURE',
            'SUPPORT_SERVICES_EXPENDITURE',
            'OTHER_EXPENDITURE',
            'CAPITAL_OUTLAY_EXPENDITURE',
            'GRADES_PK_G',
            'GRADES_KG_G',
            'GRADES_4_G',
            'GRADES_8_G',
            'GRADES_12_G',
            'GRADES_1_8_G',
            'GRADES_9_12_G',
            'GRADES_ALL_G',
            'AVG_MATH_4_SCORE',
            'AVG_READING_4_SCORE',
            'AVG_READING_8_SCORE']
target = 'AVG_MATH_8_SCORE'
```

```python
class PopulationLogReplacer(BaseEstimator, TransformerMixin):

    # All indices from ENROLL to GRADES_ALL_G refer to either Population or Money
    INDEXES = range(18) #[]

    def _init_(self):  # no *args or **kargs
        return

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        # Log for all selected indexes
        logs = []
        for i in PopulationLogReplacer.INDEXES:
            try: 
                logs.append(np.log(X[:,i]))
            except Exception as e:
                logs.append((X[:,i]))
        # Remove previous columns
        base = np.delete(X, PopulationLogReplacer.INDEXES, axis=1)
        # Insert log into base and return
        return np.c_[base, logs]

simple_imputer = SimpleImputer(strategy='median')
log_replacer = PopulationLogReplacer() ## Não será usado, por problemas não compreendidos (parece que dá uma divisão por zero?)
std_scaler = StandardScaler()

one_hot_encoder = OneHotEncoder(categories='auto', sparse = False) # state
ordinal_encoder = OrdinalEncoder() # year
```

```python
num_pipeline = Pipeline([
    ("simple_imputer", simple_imputer),
    ("poly_features", PolynomialFeatures(
            degree=5,
            include_bias=False,
    )),
    # ("log_replacer", log_replacer),
    ("std_scaler", std_scaler),
])

year_encoder = Pipeline([
    ("ordinal_encoder", ordinal_encoder),
])

state_encoder = Pipeline([
    ("one_hot_encoder", one_hot_encoder),
])
```

```python
transformer_pipeline = ColumnTransformer([
    ('numeric', num_pipeline, numeric),
    ('year', year_encoder, year),
    ('state', state_encoder, state),
])
```

```python
train_set, test_set = train_test_split(
    all_data,
    test_size=0.25,
    random_state=RANDOM_SEED,
)
X_train = train_set.drop([target], axis=1)
y_train = train_set[target]
X_test = test_set.drop([target], axis=1)
y_test = test_set[target]
```

```python
print(transformer_pipeline.fit_transform(X_train))
```

```python
lin_reg = Pipeline([
    ("transformer_pipeline", transformer_pipeline),
    ("LinearRegression", LinearRegression())
])

lasso_reg = Pipeline([
    ("transformer_pipeline", transformer_pipeline),
    ("Lasso", Lasso(1e-1))
])

res = cross_validate(
    lin_reg,
    X_train,
    y_train,
    scoring=["neg_mean_squared_error"],
    cv=5,
    n_jobs=-1,
    return_estimator=True
)

res2 = cross_validate(
    lasso_reg,
    X_train,
    y_train,
    scoring=["neg_mean_squared_error"],
    cv=5,
    n_jobs=-1,
    return_estimator=True
)
```

```python
res["test_neg_mean_squared_error"]
```

```python
res2["test_neg_mean_squared_error"]
```

Observando os erros médios quadráticos acima, é possível perceber que um modelo linear simples não foi capaz de atingir uma performance satisfatória. Sem a regularização, ele não foi capaz de se ajustar aos dados. Já o modelo com a regularização Lasso foi capaz de obter resultados ótimos, estando consistentemente com um erro médio quadrático de apenas 10 a 12.
