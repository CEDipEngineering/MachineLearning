### Tópico 1: A importância da análise exploratória
#### Interlocutor: 
#### Explicação: 
#### O objetivo de um modelo de Machine Learning é prever algo da melhor maneira possível (o preço de um objeto, se alguém possui uma doença ou não, etc) com base em dados passados, ou seja, em fatos e valores que existem ou já aconteceram antes no mundo real, por exemplo, no caso da previsão de uma doença.
#### Independente do tipo de modelo que queremos criar, um fator é constante: precisamos de um conjunto de dados para treinar esse modelo. Porém não podemos simplesmete pegar esses dados e utilizá-los cegamente em nosso modelo, pois nem sempre sabemos de que forma eles foram coletados, se possuem erros, valores faltando ou algum tipo de viés prévio.
#### Para corrigir esses possíveis problemas, que geralmente são comuns, uma vez que os dados do mundo real não são perfeitos, realizamos inicialmente uma etapa chamada análise exploratória. Como o próprio nome diz, nessa etapa, que vem antes da criação do modelo, iremos explorar o nosso conjunto de dados em busca de dados vazios, outliers e outras possíveis anormalidades.

       


### Tópico 2: Validação cruzada
#### Interlocutor: Um professor de Engenharia
#### Explicação: Suponha que estamos criando um modelo de regressão para prever o preço de uma casa dado o número de comodos, área, latitude, longitude, cor e idade. Já foi realizada a análise exploratória, estratificação dos dados, retirada de valores vazios, encoding de variáveis categóricas e separação entre treino e teste. Nos resta agora somente escolher qual modelo iremos usar, uma vez que existem várias opções para um problema de regressão, por exemplo Linear Regression, Decision Tree Regressor e Random Forest Regressor. 
#### Um ERRO CRASSO relativamente comum seria sugerir utilizar cada um desses modelos no conjunto de testes e comparar os resultados, mas essa prática é completamente errada, uma vez que o conjunto de testes só pode ser utilizado após todas decisões terem sido tomadas, garantindo assim uma avaliação não enviesada.
#### Além dessa sugestão completamente errada, outra sugestão seria utilizar o conjunto de treinamento uma vez para cada modelo e calcular a raiz do erro médio quadrático para cada um deles, e escolher o que tivesse o menor erro. Essa sugestão, apesar de não ser tão absurda quanto a anterior, não garante que esses resultados obtidos não tenham sido apenas sorte.
#### Para evitar esses problemas usamos uma técnica chamada Cross Validation, ou validação cruzada. Essa técina consiste em particionar o conjunto de treino em n partes (10 por exemplo), e calcular a raiz do erro médio quadrático para cada uma dessas partes para cada modelo. Dessa forma, podemos calcular o RMSE médio de cada modelo, além dos seus respectivos desvios padrões, escolhendo o modelo que tiver o menor erro e menor desvio padrão.





### Tópico 3: Classificação versus regressão
#### Interlocutor: Um colega no terceiro semestre do curso de Engenharia
#### Explicação:Ao criar um modelo de Machine Learning temos como objetivo, via de regra, que ele seja capaz de realizar uma previsão acertiva a partir de um conjunto de dados do passado. Dependendo do que se quer prever, ou seja, dependendo da variável dependente, esses modelos se dividem em duas categorias: Classificação ou Regressão. 
#### Um modelo de classificação é aquele no qual se quer prever uma variável dependente categórica (qualitativa), ou seja, uma variável que pode assumir um número limitado e geralmente fixo de valores/classificações. Um exemplo seria um modelo que quer prever se uma fruta é uma banana, laranja ou melancia com base em sua cor, tamanho, peso e preço por unidade. Nesse caso, nossa variável dependende pode assumir apenas 3 valores/categorias fixos: banana, laranja ou melancia, portanto trata-se de um problema de classificação. A classificação pode ser também binária, que é o caso em que a variável dependente pode assumir apenas dois valores/categorias específicos.
#### Um modelo de regressão, por outro lado, é aquel no qual se quer prever uma variável dependente contínua (quantitativa), ou seja, uma variável dependente que pode ser medida e pode assumir um número não fixo de valores. Um exemplo seria um modelo que quer prever o preço de um automóvel dado sua kilometragem, marca, cor, idade, e número de portas. Os valores possíveis da variável independente variam de 0 até o preço máximo que um carro pode atingir de forma contínua (nesse caso separados pela menor fração monetária possível).

### Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer


meu_scaler = StandardScaler()

num_pipeline = Pipeline([
    ('imputer', imputer),
    ('std_scaler', meu_scaler),
])

cat_pipeline = Pipeline([
    ('cat_encoder', encoder),
])


num_attribs = numeric = ['YEAR',
            'ENROLL',
            'TOTAL_REVENUE',
            'FEDERAL_REVENUE',
            'STATE_REVENUE',
            'LOCAL_REVENUE',
            'TOTAL_EXPENDITURE',
            'INSTRUCTION_EXPENDITURE',
            'SUPPORT_SERVICES_EXPENDITURE',
            'OTHER_EXPENDITURE',
            'CAPITAL_OUTLAY_EXPENDITURE',
            'GRADES_PK_G',
            'GRADES_KG_G',
            'GRADES_4_G',
            'GRADES_8_G',
            'GRADES_12_G',
            'GRADES_1_8_G',
            'GRADES_9_12_G',
            'GRADES_ALL_G',
            'AVG_MATH_4_SCORE',
            'AVG_READING_4_SCORE',
            'AVG_READING_8_SCORE']

cat_attribs = ['STATE']

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', cat_pipeline, cat_attribs),
])

```
